# Módulo apresentação

Módulo utilizado para apresentação dos dados. 

## Participação de mercado

```python
# Selecionando os 15 bancos com mais participação de mercado
data = pd.read_csv(ROOT_PATH + r"\dataset_kaggle.com\participacao_mercado.csv")
data = data.sort_values(by='0',ascending=True)

# Criando o gráfico de barras
plt.barh(range(len(data.head(15))), data.head(15).values[:,1])

# Adicionando os rótulos do eixo y na horizontal
plt.yticks(range(len(data.head(15))), data.head(15)['NAMEFULL'].str.split(',').str.get(0))

# Invertendo os eixos x e y para que o gráfico seja na horizontal
plt.gca().invert_xaxis()

# Adicionando um título
plt.title('Participação de Mercado (número de agências) dos 15 Bancos com Mais Participação')

# Adicionando títulos dos eixos x e y
plt.ylabel("Nome do Banco")
plt.xlabel("Participação de Mercado (%)")
```

## Análise 2.1 - Bancos por estado

```python
# Selecionando o join de bancos e estado
data = pd.read_csv(ROOT_PATH + r"\dataset_kaggle.com\visao_por_estado.csv")
# Criando o gráfico de barras
plt.bar(data.head(15).values[:,0], data.head(15).values[:,1])

# Adicionando os rótulos do eixo x na vertical
plt.xticks(range(len(data.head(15))), data.head(15).values[:,0], rotation=90)

# Adiciona título ao gráfico
plt.title('Bancos com maior número de agências por estado, 15 maiores')

# Adicionando rótulos ao eixo x e y
plt.xlabel("Estado Americano")
plt.ylabel("Quantidade de Bancos")
```

## Análise 2.2 - Ativos dos bancos por estado

```python
#Abrindo os dados em memória
top_15_estados = pd.read_csv(ROOT_PATH + r"\dataset_kaggle.com\ativos_bancos_estado.csv")

# Definido as barras do gráfico
plt.barh(range(len(top_15_estados)), top_15_estados['ASSET'])

# Adicionando os rótulos do eixo x na horizontal
plt.yticks(range(len(top_15_estados)), top_15_estados['STNAMEBR'])

# Invertendo os eixos x e y para que o gráfico seja na horizontal
plt.gca().invert_yaxis()

# Adicionando um título
plt.title('15 maiores ativos dos bancos por estado')

# Adicionando títulos dos eixos x e y
plt.ylabel('Estado')
plt.xlabel('Ativos (em trilhões de dólares)')
```

## Análise 2.3 - Comparando Ativos dos bancos com o PIB por estado

```python
# Criando um gráfico com os dados
# Para melhor visualização em formato gráfico, utilizamos somente os 10 maiores
# Exibindo somente os 10 maiores

juncao_ativos_com_pib = pd.read_csv(ROOT_PATH + r"\dataset_kaggle.com\ativos_bancos_pib_estado.csv")
ativos_pib_top_10 = juncao_ativos_com_pib.head(10)
# Configuração para apresentação de gráfico de barras agrupadas
x = np.arange(len(ativos_pib_top_10))
width = 0.40

# Definido as barras do gráfico
plt.barh(x-(width/2), ativos_pib_top_10['ASSET_%'], width, color='blue', label='% Ativo')
plt.barh(x+(width/2), ativos_pib_top_10['2021gdp_%'], width, color='green', label='% PIB')

# Obtendo nome dos estados pela coluna 'STNAMEBR'
lista_estados = ativos_pib_top_10['STNAMEBR']

# Definindo os rótulos do eixo x com os nomes dos estados
plt.yticks(x, lista_estados, fontsize=8)

# Invertendo os eixos x e y para que o gráfico seja na horizontal
plt.gca().invert_yaxis()

# Adicionando um título
plt.title('Relação entre Ativos do Banco com o PIB por estado')

# Adicionando títulos dos eixos x e y
plt.ylabel('Estado EUA')
plt.xlabel('Participação (%)')

# Adicionando as legendas das barras
plt.legend()

# Mostrando o gráfico
plt.show()
```
